```python
labels = ['% of Training Dataset Used', '% of Images Annotated', 'DICE Score']
proposed_means = [0.01, 0, 0.9114]
sota_means = [1.0, 0.023, 0.952]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, proposed_means, width, label='Proposed')
rects2 = ax.bar(x + width/2, sota_means, width, label='SOTA')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
